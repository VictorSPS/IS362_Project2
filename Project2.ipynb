```python
import numpy as np
import pandas as pd
import decimal
import os
```

I imported csv file with data set of professional CSGO players' mouse sensitivities, that was provided by Kenny Lee on the discussion board.

```python
data = pd.read_csv("CS_GO_untide_data.csv")
data
```

Each player uses different senstivities, which is defined by three parts: DPI, in-game sensitivity, and windows sensitivity. For better visability of main parameters of sensativity I decided to create 3 new tables.



```python
DPI = data[['CC', 'Player name', 'Team name', 'DPI']]
in_game_sensitivity = data[['CC', 'Player name', 'Team name', 'sensitivity']]
Windows_sensitivity = data[['CC', 'Player name', 'Team name', 'Windows sensitivity']]
List = [DPI, in_game_sensitivity, Windows_sensitivity]


print('DPI sensitivity:')
List[0].sort_values('CC', ascending=True).head(10)
```

```python
print('in-game sensitivity:')
List[1].sort_values('CC', ascending=True).head(10)
```

```python
print('Windows sensitivity:')
List[2].sort_values('CC', ascending=True).head(10)
```

Then I used melting to visualize main parameters for each player in one table.

```python
new_data = pd.melt(data,
                   id_vars=['CC', 'Player name'],
                   value_vars=['DPI', 'sensitivity', 'Windows sensitivity'])
new_data = new_data.sort_values(by=["Player name"])
new_data.head(20)
```
